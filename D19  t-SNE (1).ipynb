```python
from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
```

```python
# load data
filename = 'TSNE_data.csv'
dataframe = pd.read_csv(filename)
dataframe

# diagnosis:  ఒక నిర్దిష్ట అనారోగ్యం లేదా సమస్య ఏమిటో గుర్తించడం
```

```python
dataframe.info()
```

```python
# Split-out validation dataset
array = dataframe.values

# separate array into input and output components
X = array[:,1:]
Y = array[:,0]       # 1st col
```

```python
#TSNE visualization
from bioinfokit.visuz import cluster

data_tsne = TSNE(n_components=2,perplexity=50).fit_transform(X)   #Reduce 30 to  2column
cluster.tsneplot(score=data_tsne)1


# t-SNE is particularly useful for visualizing high-dimensional data in a lower-dimensional space (typically 2D or 3D)
# from bioinfokit.visuz import cluster:
# perplexity=50 more perplexity More accept the data points
# perplexity=50   dhagera ga points , Increase perplexity=500 Mari dhagera ga image
# .fit_transform(X): This part applies the t-SNE algorithm to the input data X. It first "fits" the algorithm to the data, learning the underlying structure, and then "transforms" the data into the lower-dimensional space (2D in this case). The result is stored in the data_tsne variable.
```

```python
# get a list of categories
color_class = dataframe['diagnosis'].to_numpy()
color_class
```

```python
cluster.tsneplot(score=data_tsne, colorlist=color_class, legendpos='upper right',legendanchor=(1.15, 1))


#Plot will be stored in the default directory
# to_numpy()--->This converts the selected column (which is a pandas Series) into a NumPy array
# legendpos='upper right': This specifies the position of the legend on the plot (upper right corner)
# legendanchor=(1.15, 1): This further refines the legend's position, allowing for fine-tuning its placement relative to the plot's boundaries.
# legendanchor=(1.15, 1)) ---> right 1.5 , up 1
```
